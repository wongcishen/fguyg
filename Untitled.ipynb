```python
#image 250 x 256

img=np.fromfunction(lambda i,j:j,(50,256),dtype=np.uint8)

#Trackback func
def trackbar(x):
    ret,th1=cv.threshold(img,x,255,cv.THRESH_BINARY)
    ret,th2=cv.threshold(img,x,255,cv.THRESH_BINARY_INV)
    ret,th3=cv.threshold(img,x,255,cv.THRESH_TRUNC)
    ret,th4=cv.threshold(img,x,255,cv.THRESH_TOZERO)
    ret,th5=cv.threshold(img,x,255,cv.THRESH_TOZERO_INV)
    
    cv.imshow('hello',np.vstack([th1,th2,th3,th4,th5]))
    
cv.namedWindow('window')
cv.createTrackbar('threshold','window',100,255,trackbar)

cv.waitKey(0)
cv.destroyAllWindows()

```
